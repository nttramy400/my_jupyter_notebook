### Correlation coefficient (Pearson's)

scipy.stats.pearsonr(x, y)

```python
import numpy as np
from scipy.stats import pearsonr
a = np.array([45,63,36,52,29,33,48,55,72,66])
b = np.array([20,19,16,20,12,14,16,18,20,22])
n = len(a)
cor1 = (sum(a*b) - n*np.mean(a)*np.mean(b))/(n*np.std(a)*np.std(b))
cor2 = pearsonr(a,b)
print (cor1)#manual
print (cor2)#sklearn
```
